```python
# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS as stopwords
```

```python
# 소설 'the old man and the sea' 를 불러옵니다.
old_text = open("/content/drive/MyDrive/21-1/theoldmanandthesea.txt","r").read()
old_text
```

```python
# 텍스트에서 유의미한 단어 토큰만을 선별하기 위해서 큰 의미가 없는 단어 토큰을 제거하는 작업이 필요합니다.
# 다음과 같이 문장에서 빈도수는 높지만 분석에 쓸모없는 단어들을 stopwords(불용어)라 합니다. 
stopwords
```

```python
# old text의 워드클라우드를 만드는 WordCloud 클래스의 인스턴스 생성
old_wc = WordCloud(background_color = "white",
                  collocations = False, 
# 이때 collocations=True 이면 said King, said Hatter 같이 자주 나타나는 단어는 하나의 어구로 분류됨.
                  stopwords = stopwords,
                   width = 600, height = 400, colormap = "flag",
                   min_font_size = 5) 
# alice_novel이라는 텍스트 데이터를 사용하여 wordcloud를 generate하기
old_wc.generate(old_text)
```

```python
# print frequency of each words
# 가장 많이 나오는 단어가 1 이 되고 나머지가 상대적인 값으로 나타남.
old_wc.words_
```

```python
#visualization of wordcloud
old_wc.to_image()
```
